```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sample_bank_statement.csv")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df.set_index("Date", inplace = True)
df.head()
```

```python
df["Amount"] = df["Amount"].str.replace(",", "", regex=False).astype(float)
df.head()
```

```python
monthly_exp = df.resample("ME")["Amount"].sum()
print(f"Monthly Expense is: ")
monthly_exp
```
